### ***Columnas de valor*** ###
* Record :  Numero de grabado
* Date : Fecha de grabado
* LOC_Time : Hora del grabado
* Articulo : Articulo analizado
* Estado Camara : Boleano indicador de funcionamiento
* Comunicacion Chronos : Boleando indicador de comunicación con IT
* Estado Controles : Boleano indicar de habilitación de controles
* Resultado Estampado : Resultado de la coincidencia del Articulo con el número estampado
* Resultado Controles: Resultado de la coincidencia de los parametros del dentado

```python
Record = []
Date = []
LOC_Time = []
Articulo = []
Statecamera = []
Comunicacion_Chronos = []
Statecontrol = []
Statestamp = list(np.random.binomial(1,0.7383,len(df)//2))
Resultcontrol = list(np.random.binomial(1,0.82,len(df)//2))

Articulotrabajado = 2829086



for i in range(0, len(df), 2):
    if i < len(df):
        Record.append(df.iloc[i]["Record"])
        Date.append(df.iloc[i]["Date"])
        LOC_Time.append(df.iloc[i]["LOC Time"])
        Articulo.append(Articulotrabajado)
        Statecamera.append(1)
        Comunicacion_Chronos.append(1)
        Statecontrol.append(1)
        # Statestamp.append(df.iloc[i]["Estado Estampado"])
        
        # # Agregamos Resultcontrol solo si el siguiente índice existe
        # if i + 1 < len(df):
        #     Resultcontrol.append(df.iloc[i + 1]["Estado Estampado"])
        # else:
        #     Resultcontrol.append(0)  # O cualquier otro valor por defecto
```

```python
data = {
    'Record': Record,
    'Date': Date,
    'LOC Time': LOC_Time,
    'Articulo': Articulo,
    'Estado Camara': Statecamera,
    'Comunicacion Chronos': Comunicacion_Chronos,
    'Estado Controles': Statecontrol,
    'Resultado Estampado': Statestamp,
    'Resultado Controles': Resultcontrol
}

df_new = pd.DataFrame(data)
df_new.head(10)
```
